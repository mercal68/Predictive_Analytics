```python
# import Keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
import numpy as np
import pandas as pd
seed = 7
np.random.seed(seed)
```

```python
# Data is obtained at https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes
dataset=pd.read_csv("pima.csv", header=None)
```

```python
X=dataset.iloc[:, 0:8]
y=dataset.iloc[:, -1]
```

```python
#Normalize & scale features
from sklearn import preprocessing as pre
scaler = pre.StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
from keras.layers import Dense
```

```python
# create first model 
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X_scaled, y, nb_epoch=100, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X_scaled, y)
```

```python
# obtain scores
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
#create 2nd model with added layers
model = Sequential()
model.add(Dense(50, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(50, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the new model
model.fit(X_scaled, y, nb_epoch=100, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X_scaled, y)
```

```python
# obtain scores
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
